Create a program that simulates the allelic frequency in a finite diploid population for a certain number of generations.

```python
#Packages
from numpy import random as rd
from plotly.graph_objs import *

#Allele frequencies
initA = 0.50
inita = 0.50

#Fitnesses
fAA = 1
fAa = 1
faa = 1

#Pop Size
pop = 1000

#Number of generations
gen = 100

#Number of simulations
sim = 100
```

```python
import numpy as np
import plotly.express as px
import plotly.graph_objs as go

def calcNextGen(freqA, fitness, pop):
    freqa = 1 - freqA
    fAA = freqA ** 2
    fAa = 2 * freqA * freqa
    faa = freqa ** 2
    # wAA = fAA * pop
    # wAa = fAa * pop
    # waa = faa * pop
    wAA = fitness[0]
    wAa = fitness[1]
    waa = fitness[2]
    M = wAA*fAA + wAa*fAa + waa*faa
    fAA1 = (wAA*fAA) / M
    fAa1 = (wAa*fAa) / M
    faa1 = (waa*faa) / M
    nAA1 = np.random.binomial(pop, fAA1)
    nAa1 = np.random.binomial(pop, fAa1)
    naa1 = pop - nAA1 - nAa1
    # fA1 = (np.random.binomial(pop, nAA1)) + ((np.random.binomial(pop, nAa1))/2)
    fA1 = (nAA1 + (1/2) * nAa1) / pop
    return fA1

def simulate(initA, pop, fitness, gen, sim):
    A_vals = []
    nextA = calcNextGen(initA, fitness, pop)
    A_vals.append(initA)
    fig = go.Figure()
    a = np.zeros(shape=(sim, gen+1))
    b = np.array(range(gen+1))
    hist = []
    for z in range(sim):
        for i in range(gen):
            A_vals.append(nextA)
            nextA = calcNextGen(initA, fitness, pop)
        fig.add_trace(go.Scatter(x = b, y = A_vals, mode = 'lines')) 
        hist.append(A_vals[-1])                     
        a[z] = A_vals
        print(A_vals)
        A_vals = []
        A_vals.append(initA)

    fig2 = px.histogram(x= hist)
    fig2.show()
    #print(a)
    #fig = px.line(x = b, y = a)
    fig.show()

simulate(.5, 100, [1,1,1], 100, 100)
```
